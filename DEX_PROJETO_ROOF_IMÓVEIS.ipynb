<h1><b><u>CONDITION - CONDIÇÃO DO IMÓVEL</u></b></h1><br>
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b7/Seattle_-_Craftsman_revival_houses.jpg/800px-Seattle_-_Craftsman_revival_houses.jpg?20090518050858">
<br><br>
<div align='justify'><h5>Esta variável representa a condição atual do imóvel, em uma escala de 1 a 5. É geralmente definida como:</h></div><br>
<div align='justify'><h5><i><b>1- Ruim - Desgastado.</b> É necessário reparação e revisão nas superfícies pintadas, telhados, canalizações, aquecimento e há inúmeras inadequações funcionais. Manutenção muito adiada, abusos na utilização, com valor de uso limitado, próximo do abandono ou necessidade de grande reforma; reutilização ou mudança na ocupação é iminente. A idade efetiva está próxima do final da escala, independentemente da idade cronológica real.

<b>2 - Razoável - Mal gasto.</b>  Muito reparo necessário. Muitos itens precisam de repintura ou revisão, é evidente que há manutenção pendente, as utilidades e os sistemas de construção estão inadequados, todos encurtando a expectativa de vida útil do imóvel e aumentando a sua idade efetiva.

<b>3 - Média.</b> Alguma evidência de manutenção adiada e obsolescência normal com a idade, em que são necessários alguns pequenos reparos, juntamente com alguns retoques. Todos os principais componentes ainda estão funcionais e contribuindo para uma expectativa de vida útil estendida. A idade efetiva e as utilidades são padrão para propriedades semelhantes.

<b>4 - Boa.</b> Nenhuma manutenção aparente é necessária, mas nem tudo é novo. A aparência e as utilidades estão acima do padrão, e a idade efetiva geral é menor do que as outras propriedades similares.

<b>5 - Muito Boa.</b> Todos os itens estão bem conservados, muitos tendo sido renovados ​​e reparados pois apresentaram sinais de desgaste, aumentando a vida útil e diminuindo a idade efetiva do imóvel, com pouca deterioração ou obsolescência evidente, e com as utilidades em alto nível.</h5></i></div>

<h1><b><u>MATRIZ DE CORRELAÇÃO DAS VARIÁVEIS</u></b></h1>

<div align='justify'><h5><i>Nesta seção será apresentada a matriz de correlação do dataset, para visualizar a associação entre as fariáveis do conjunto de análise. A matriz de correlação, visualmente transformada em um mapa de calor <b>(heatmap)</b> fornece uma visualização clara das variáveis e de suas interrelações.<br>

Cada variável (<b>feature</b>) é listada em ambos eixos, e suas relações com as outras variáveis são apresentadas utilizando um gradiente de cores. Na medida em que a cor muda (em ambas direções), isso significa que as variáveis são altamente correlacionadas.<br>

<h1><b><u>LOCALIZAÇÃO DOS IMÓVEIS POR PREÇO</u></b></h1>


<div align = 'justify'><h5><i>No mapa abaixo, é possível visualizar geograficamente a localização dos imóveis pelo seu preço, sendo as cores mais próximas do azul os de menor valor, e as cores mais próximas de vermelho as de maio valor.</i></h5></div>



```python
cmap = cm.LinearColormap(colors=['blue','yellow','red'],vmin=100000,vmax=3000000) #escolha das cores do mapa
m = folium.Map(location=[df.lat.mean(),df.long.mean()],tiles='cartodbpositron',zoom_start=10)
for i in range(len(df)):
     folium.Circle(
         location=[df.iloc[i]['lat'], df.iloc[i]['long']],
         radius=10,
         fill=True,
         color=cmap(df.iloc[i]['price']),
         fill_opacity=0.7
     ).add_to(m)
m.add_child(cmap)
```
